```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}
fields = {'BusinessName':1, 'BusinessType':1, 'LocalAuthorityName':1, 'RatingValue': 1, 'scores.Hygiene' : 1, 'scores.Structural' : 1, 'scores.ConfidenceInManagement' : 1, 'geocode.longitude':1, 'geocode.latitude':1}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f'Number of establishments with a hygiene score of 20: {count}')

# Display the first document in the results using pprint
results = establishments.find_one(query, fields)
print("First document with a hygiene score of 20:")
pprint(results or "No documents found with a hygiene score of 20.")
```

```python
# Convert the result to a Pandas DataFrame
hygiene_doc = list(establishments.find(query, fields))
hygiene_df = pd.DataFrame(hygiene_doc)
# Display the number of rows in the DataFrame
rows = hygiene_df.shape[0]
print(f'Number of rows: {rows}')
# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'LocalAuthorityName': {'$regex': '.*London.*'}, 'RatingValue': {'$gte': 4}}
fields = {'BusinessName':1, 'BusinessType':1, 'LocalAuthorityName':1, 'RatingValue': 1, 'scores.Hygiene' : 1, 'scores.Structural' : 1, 'scores.ConfidenceInManagement' : 1, 'geocode.longitude':1, 'geocode.latitude':1}                                                  

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f'London establishments with a rating >= 4: {count}')
# Display the first document in the results using pprint
results = establishments.find_one(query, fields)
print("First document of London establishment with a rating >= 4:")
pprint(results)
```

```python
# Convert the result to a Pandas DataFrame
london_rating_doc = list(establishments.find(query, fields))
london_rating_df = pd.DataFrame(london_rating_doc)

# Display the number of rows in the DataFrame
rows = london_rating_df.shape[0]
print(f'Number of rows: {rows}')

# Display the first 10 rows of the DataFrame
london_rating_df.head(10)
```
